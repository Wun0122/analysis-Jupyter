array的選項可以利用判斷來挑選元素

array中符合元素條件的array = array[條件]

```python
x = np.linspace(-5, 5, 500)
y = np.sinc(x)

plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.plot(x[y>0] ,y[y>0]),'o'
```
